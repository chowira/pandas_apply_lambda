```python
 work_df = pd.read_csv("../data/input/IMDB-Movie-Data.csv")
work_df.head()
```

```python
#funcion challenger 1
def create_bin(x):
    if x in range(0,1000):
        return "cat 1"
    elif x in range(999,10000):
        return  "cat 2"
    elif x in range(9999,100000):
        return "cat 3"
    elif  x in range(99999,100000):
        return "cat 4"
    elif x >= 1000000:
        return "cat 5"
```

```python
create_bin(999)
```

```python
work_df["bin"] = work_df["Votes"].apply(create_bin)
work_df
```

```python
cost_for_minutes = work_df["Runtime (Minutes)"] / work_df["Revenue (Millions)"]
cost_for_minutes
```

```python
lambda row :row(["Runtime (Minutes)"]/work_df["Revenue (Millions)"], axis = 1) 
```
